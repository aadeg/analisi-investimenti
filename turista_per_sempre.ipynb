# Rendimento Turista per Sempre
Download and cleaning dei rendimenti del gratta e vinci Turista per Sempre per analisi [Misure di Rischio](./misure_rischio.ipynb).

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
sns.set_style("whitegrid")
# sns.set_theme()
matplotlib.rcParams['figure.figsize'] = [9.0/2, 5.0/2]
```

```python
qt_tot = 54720000
df = pd.read_html("https://www.adm.gov.it/portale/-/new-turista-per-sempre",
    decimal=',', thousands='.')
df = df[0]
df
```

```python
df.info()
```

```python
df_clean = df.rename(columns={
    df.columns[0]: "vincita",
    df.columns[1]: "qt",
    df.columns[2]: "1_qt"})
df_clean.vincita = pd.to_numeric(df_clean.vincita, errors="coerce")
df_clean.qt = pd.to_numeric(df_clean.qt, errors="coerce")
df_clean["1_qt"] = pd.to_numeric(df_clean["1_qt"], errors="coerce")

# df_clean = df_clean.drop(13)
# df_clean = pd.concat([df_clean, pd.DataFrame([0])], axis=0, ignore_index=True)
df_clean.loc[13, "vincita"] = 0
df_clean.loc[13, "qt"] = qt_tot - df_clean.loc[0:12, "qt"].sum()
df_clean["1_qt"] =  qt_tot / df_clean.qt
df_clean["prob"] = 1 / df_clean["1_qt"]
df_clean["prob cum"] = df_clean.prob.cumsum()
df_clean
```
